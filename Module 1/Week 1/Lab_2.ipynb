## importing modules


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## creating data set into `x_train` and `y_train`

```python
#x_train
#y_train
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

## for training samples! (m)

```python
m = len(x_train)
print(f"training samples for our dataset is {m}")
```

## for taking (x_i, y_i)

```python
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^({i}) = ({x_i}, {y_i})")
```

## ploting the scatter plot

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing prices")
plt.xlabel("area")
plt.ylabel("price in 1000s$")
plt.show()
```
